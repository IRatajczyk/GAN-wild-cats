```python
import os
import cv2
```

```python
SOURCE: str = r"C:\Users\igor\Desktop\Studia\I\Semestr6\DNN\Lab4\archive\cats_projekat"
DESTINATION: str = r"C:\Users\igor\Desktop\Studia\I\Semestr6\DNN\Lab4\processed"
```

```python
# open every file in every subdirectory of PATH and resize it to 64x64

for subdir in os.listdir(SOURCE):
    for file in os.listdir(os.path.join(SOURCE, subdir)):
        print(os.path.join(SOURCE, subdir, file))
        img = cv2.imread(os.path.join(SOURCE, subdir, file))
        img = cv2.resize(img, (64, 64))
        cv2.imwrite(os.path.join(DESTINATION, file), img)

```
